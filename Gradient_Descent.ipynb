```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000 #increase once you set the optimum learning rate
    n = len(x) #x and y length assumed to be same
    learning_rate = 0.01#0.08 #Cost function should be reducing
    plt.scatter(x,y,color='red',marker='+',linewidths=4)
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr   #y = mx +b
        plt.plot(x,y_predicted,color='green')
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - md * learning_rate
        b_curr = b_curr - bd * learning_rate
        
        #plt.scatter(x,y_predicted,color = 'blue')
        #plt.plot(x,y_predicted,color = 'red')
        print("m {},b {},iteration {},cost {}".format(m_curr,b_curr,i,cost))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
