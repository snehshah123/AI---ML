```python
import pandas as pd
df = pd.read_csv(r"G:\SRAV2\py-master\py-master\ML\6_train_test_split\carprices.csv")
df.head()  
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Car Mileage")
plt.ylabel("Sell Price in Dollars")
plt.scatter(df['Mileage'],df['Sell Price($)'])
plt.tight_layout()
```

```python
x = df[['Mileage','Age(yrs)']]
y = df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2) #random_state = 10 for same values everytime 
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
y_test  
```

```python
reg.score(x_test,y_test)
```
